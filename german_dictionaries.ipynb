```python
##lib importing
import pandas as pd
```

```python
##files loading
path = "datasets/"
filename = "german.csv"
german = pd.read_csv(path + filename, index_col=0)
german
```

```python
german = german.rename(columns={"vehicleType": "vehicle_type", "yearOfRegistration": "year", "gearbox": "transmission", "kilometer": "km", "fuelType": "fuel_type"})
```

```python
set(german['vehicle_type'])
```

```python
vehicle_type_dict = {"andere" : "other", 'bus' : "van", 'cabrio' : "cabrio", 'coupe' : "coupé", 'kleinwagen' : "small_car", 'kombi' : "sports_tourism_van", 'limousine' : 'large_family_car', 'suv' : 'suv'}
```

```python
german['vehicle_type'] = german['vehicle_type'].map(vehicle_type_dict)
```

```python
transmission_dict = {"manuell": "manual", "automatik": "automatic"}
german['transmission'] = german['transmission'].map(transmission_dict)

german
```

```python
german['horsepower'] = german['horsepower'].astype(int)
```

```python
german['fuel_type'].unique()
```

```python
fuel_dict = {"benzin": "gas", "diesel": "diesel", "lpg": "GPL", 'andere': "other", 'hybrid': 'hybrid', 'cng': "CNG"}
```

```python
german['fuel_type'] = german['fuel_type'].map(fuel_dict)
```

```python
german.dtypes
```

```python
german.to_csv(path + "german_transl.csv")
german
```
