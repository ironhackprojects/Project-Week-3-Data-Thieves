```python
import requests
import urllib.request
import time
import re
import pandas as pd
import numpy as np
import collections
import random
from bs4 import BeautifulSoup
from time import sleep
```

```python
#Global Variable definition

ad_link = 'https://www.standvirtual.com/anuncio/' #header of car adds to scrape

# Data to keep after scraping per car
keep_columns = ('Ano de Registo',  'Classe do veículo', 'Combustível','Condição', 'Marca',  'Matrícula', 'Origem', 'Potência', 'Quilómetros', 'Segmento', 'Série', 'Tipo de Caixa', 'Valor Fixo', 'ID')
```

```python
#Generate inks of pages to scrape

page_list = []

for page in range(500):
    url_s = 'https://www.standvirtual.com/carros/?search%5Border%5D=created_at%3Adesc' # base link - this link may need to be updated
    new_url = str(url_s+'&page='+str(page))
    page_list.append(new_url)
    
print(page_list)
```

```python
#Scraping

all_dics = []
count_iter = 0
page_list2 = page_list[::-1]

for page in page_list2:

    print(page)
    response = requests.get(page)
    soup = BeautifulSoup(response.text, 'html.parser')
    #soup_list.append(soup)

    scrap = soup.find_all('a', href = True)
    links = [tag['href'] for tag in scrap]
    links3 = list(set([el for el in links if ad_link in el]))
    
    for el in links3:
      soup2 = BeautifulSoup(requests.get(el).text, 'html.parser')
      scrap2 = soup2.find_all('li', class_ = 'offer-params__item')
      scrap3 = soup2.find_all('span', class_ = 'offer-meta__value')
      car_id = [el.text for el in scrap3]
      
      if len(car_id) == 0 : 
            print(el)
            continue
                 
      car_id = [el.text for el in scrap3][-1]
        
      list_raw2 = [re.sub(r'  +',':',el.text.strip().replace('\n','')) for el in scrap2]
      dic_raw2 = {el.split(':')[0]: el.split(':')[1] for el in list_raw2}
      dic_raw2.update({'ID': car_id})
    
      keys = list(dic_raw2.keys())      
      for key in keys:
          if key not in keep_columns:
             dic_raw2.pop(key,None)
      
      all_dics.append(dic_raw2)
        
    all_dics_ordered =[collections.OrderedDict(sorted(dic.items())) for dic in all_dics]
      
    df = pd.DataFrame(all_dics_ordered)
      
    df_filename = 'StdV' + str(count_iter) + ".csv"
      
    count_iter += 1
      
    pd.DataFrame.to_csv(df, df_filename)
        
    sleep (random.random())
```
