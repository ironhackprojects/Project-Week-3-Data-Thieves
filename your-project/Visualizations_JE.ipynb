```python
import pandas as pd
import matplotlib.pyplot as plt
from mat
import pandas as pd
import seaborn as sns
import numpy as np
sns.set(palette = 'Pastel1')
%matplotlib inline
```

```python
pt_data = pd.read_csv('Full_SV.csv',index_col=[0])
deutsch_data = pd.read_csv('german_transl.csv', index_col=[0])
```

```python
pt_data.drop('ID', axis = 1,inplace= True)
pt_data
```

```python
deutsch_data
```

```python
hp_pt = pt_data.groupby(['year']).mean() [['horsepower']]
hp_deut = deutsch_data.groupby(['year']).mean() [['horsepower']]
hp_pt
```

```python
fig_dims = (12, 6)
fig, ax = plt.subplots(figsize=fig_dims)

x = sns.lineplot(hp_pt.index[46:60], hp_pt['horsepower'][46:60],label = 'Portugal')
y = sns.lineplot(hp_deut.index[71:],hp_deut['horsepower'][71:], label = 'Germany')
plt.setp(x.lines,linewidth=3)  

plt.setp(ax.get_legend().get_texts(), fontsize='25')
plt.setp(ax.get_legend().get_title(), fontsize='30')

leg = plt.legend()
leg.get_frame().set_linewidth(0.0)

plt.title('Average Horse Power').get_figure().savefig('Average Horse Power.png', dpi=500)
```

```python
fig_dims = (12, 6)
fig, ax = plt.subplots(figsize=fig_dims)

#ns.scatterplot(x="CO2 Emission Grams/Mile", y="Highway MPG", hue="Drivetrain", data=vehicles)
x = sns.scatterplot(pt_data['year'], pt_data['horsepower'],label = 'Portugal', linewidth = 0)
y = sns.scatterplot(deutsch_data['year'], deutsch_data['horsepower'], label = 'Germany', linewidth = 0)
plt.setp(x.lines,linewidth=3)  

plt.setp(ax.get_legend().get_texts(), fontsize='25')
plt.setp(ax.get_legend().get_title(), fontsize='30')

plt.xlim(2005, 2020)

leg = plt.legend()
leg.get_frame().set_linewidth(0.0)
```

```python
hp_pt_br = pt_data.groupby(['year','brand']).mean() [['horsepower']]
hp_deut_br = deutsch_data.groupby(['year','brand']).mean() [['horsepower']]

five_brands = ['BMW', 'Mercedes-Benz', 'Renault', 'Peugeot', 'VW']
#Ser_pt = hp_pt_br[np.in1d(hp_pt_br.index.get_level_values(1), five_brands)]
Ser_pt = pd.pivot_table(hp_pt_br, values='horsepower', index=['year'], columns=['brand'], aggfunc=np.mean)
Ser_pt = Ser_pt.loc[2005:2019,five_brands]
Ser_pt

five_brands_d = ['volkswagen', 'bmw', 'audi','mercedes_benz', 'opel']
Ser_d = pd.pivot_table(hp_deut_br, values='horsepower', index=['year'], columns=['brand'], aggfunc=np.mean)
Ser_d= Ser_d.loc[2005:2019,five_brands_d]
```

```python
fig_dims = (12, 6)
fig, ax = plt.subplots(figsize=fig_dims)

#x = sns.lineplot(data = Ser_pt, linestyle="-",lw = 5)
x = sns.lineplot(Ser_pt.index, Ser_pt['BMW'], label = 'BMW')
y = sns.lineplot(Ser_pt.index, Ser_pt['Mercedes-Benz'], label = 'Mercedes-Benz')
z = sns.lineplot(Ser_pt.index, Ser_pt['Renault'], label = 'Renault', color = 'b', alpha = 0.5)
r = sns.lineplot(Ser_pt.index, Ser_pt['Peugeot'], label = 'Peugeot', color = 'r', alpha =0.5)
s = sns.lineplot(Ser_pt.index, Ser_pt['VW'],  label = 'VW',color = 'g', alpha =0.4)
plt.setp(x.lines,linewidth=3)  
plt.ylabel('horsepower')

leg = plt.legend()
leg.get_frame().set_linewidth(0.0)
plt.setp(ax.get_legend().get_texts(), fontsize='12')
plt.setp(ax.get_legend().get_title(), fontsize='20')
plt.title('Top 5 Brands Average Horse Power - Portugal').get_figure().savefig('Top5_Average_Horse_Power_PT.png', dpi=500)
```
